# What level of total dollar amount has every artist earned?
• 3 Levels:
• income > 50 High, 50 >= income > 20 Medium, income <= 20, Low

```python
%%sql

SELECT *
    FROM "Artist";
```

```python
%%sql

SELECT (regexp_split_to_table(R."Name", E'[/,&]')) AS "ArtistName", 
    SUM(IL."UnitPrice" * IL."Quantity") AS "DollarAmountEarned",
    CASE 
        WHEN SUM(IL."UnitPrice" * IL."Quantity") > 50 THEN 'High'
        WHEN SUM(IL."UnitPrice" * IL."Quantity") > 20 THEN 'Medium'
        ELSE 'Low'
    END AS "LevelOfDollarAmountEarned"
    FROM "Artist" R
    JOIN "Album" A
    ON R."ArtistId" = A."ArtistId"
    JOIN "Track" T
    ON A."AlbumId" = T."AlbumId"
    JOIN "InvoiceLine" IL
    ON T."TrackId" = IL."TrackId"
    GROUP BY R."Name"
    ORDER BY SUM(IL."UnitPrice" * IL."Quantity") DESC;
```

Answer: See the table above.

# Who are the top 10 customers who spent most on U2?

```python
%%sql

SELECT C."FirstName", C."LastName", SUM(IL."UnitPrice" * IL."Quantity") AS "AmountSpent"
    FROM "Customer" C
    JOIN "Invoice" I
    ON C."CustomerId" = I."CustomerId"
    JOIN "InvoiceLine" IL
    ON I."InvoiceId" = IL."InvoiceId"
    JOIN "Track" T
    ON IL."TrackId" = T."TrackId"
    JOIN "Album" A
    ON T."AlbumId" = A."AlbumId"
    JOIN "Artist" R
    ON A."ArtistId" = R."ArtistId"
    WHERE R."Name" LIKE '%U2%'
    GROUP BY C."CustomerId"
    ORDER BY "AmountSpent" DESC
    LIMIT 10;
    
```

Answer: See the table above.

# Find all songs(tracks) by Pink Floyd(Artist)

```python
%%sql

SELECT T."Name" AS "Songs by Pink Floyd"
    FROM "Artist" R
    JOIN "Album" A
    ON R."ArtistId" = A."ArtistId"
    JOIN "Track" T
    ON A."AlbumId" = T."AlbumId"
    WHERE R."Name" LIKE '%Pink Floyd%';
```

#  Find all Rock songs performed by any artist whose name starts with a 'D'

```python
%%sql

CREATE OR REPLACE view "view_RockSongs" AS

SELECT T."Name" AS "RockSongs", (regexp_split_to_table(R."Name", E'[/,&]')) AS "ArtistName"
    FROM "Track" T
    JOIN "Genre" G
    ON T."GenreId" = G."GenreId"
    JOIN "Album" A
    ON T."AlbumId" = A."AlbumId"
    JOIN "Artist" R
    ON A."ArtistId" = R."ArtistId"
    WHERE G."Name" = 'Rock';
```

```python
%%sql

SELECT "RockSongs", "ArtistName"
    FROM "view_RockSongs"
    WHERE "ArtistName" LIKE 'D%';
```

Answer: See the table above.

```python
%%sql

DROP VIEW IF EXISTS "view_RockSongs";
```

```python
%%sql

SELECT "RockSongs", "ArtistName"
    FROM "view_RockSongs"
    WHERE "ArtistName" LIKE 'D%';
```

# What genres of songs are performed in the album called 'Morning Dance'?

```python
%%sql

SELECT DISTINCT G."Name" AS "GenreName"
    FROM "Genre" G
    JOIN "Track" T
    ON G."GenreId" = T."GenreId"
    JOIN "Album" A
    ON T."AlbumId" = A."AlbumId"
    WHERE A."Title" = 'Morning Dance';
```

Answer: Jazz

#  Which artists have performed in songs of more than one genre?

```python
%%sql

CREATE OR REPLACE view "view_GenresOfArtists" AS

SELECT DISTINCT G."Name" AS "GenreName", (regexp_split_to_table(R."Name", E'[/,&]')) AS "ArtistName"
    FROM "Genre" G
    JOIN "Track" T
    ON G."GenreId" = T."GenreId"
    JOIN "Album" A
    ON T."AlbumId" = A."AlbumId"
    JOIN "Artist" R
    ON A."ArtistId" = R."ArtistId"
    ORDER BY "ArtistName";
```

```python
%%sql

SELECT *
    FROM "view_GenresOfArtists"
```

```python
%%sql

SELECT "ArtistName", COUNT("GenreName") AS "NumOfGenre"
    FROM "view_GenresOfArtists"
    GROUP BY "ArtistName"
    HAVING COUNT("GenreName") > 1;
```

Answer: See the table above.

```python
%%sql

DROP VIEW IF EXISTS "view_GenresOfArtists";
```

```python
%%sql

SELECT "ArtistName", COUNT("GenreName") AS "NumOfGenre"
    FROM "view_GenresOfArtists"
    GROUP BY "ArtistName"
    HAVING COUNT("GenreName") > 1;
```

# Which composers have written songs for more than one album?
• (For this query, you have to report individual composers) Hint: Use regular expressions (regexp functions). Here is
an example https://stackoverflow.com/questions/35634387/postgresql-regex-split-column-to-array (Links to an external
site.)

```python
%%sql

SELECT regexp_split_to_table("Composer", E'[&/,]') AS "ComposerName", COUNT("AlbumId")
    FROM "Track"
    GROUP BY "ComposerName"
    HAVING COUNT("AlbumId") > 1
    ORDER BY COUNT("AlbumId");
    
```

Answer: See the table above.

# Find the top 5 artists based on the sales of their albums.

```python
%%sql

SELECT (regexp_split_to_table(R."Name", E'[/,&]')) AS "ArtistName",
    SUM(IL."UnitPrice" * IL."Quantity") AS "AlbumSales"
    FROM "InvoiceLine" IL
    JOIN "Track" T
    ON IL."TrackId" = T."TrackId"
    JOIN "Album" A
    ON T."AlbumId" = A."AlbumId"
    JOIN "Artist" R
    ON A."ArtistId" = R."ArtistId"
    GROUP BY "ArtistName"
    ORDER BY "AlbumSales" DESC
    LIMIT 5;
```

Answer: See the table above.

#  Find the ID and name of those customers who have purchased Jazz music.

```python
%%sql

SELECT C."FirstName", C."LastName", C."CustomerId"
    FROM "Customer" C
    JOIN "Invoice" I
    ON C."CustomerId" = I."CustomerId"
    JOIN "InvoiceLine" IL
    ON I."InvoiceId" = IL."InvoiceId"
    JOIN "Track" T
    ON IL."TrackId" = T."TrackId"
    JOIN "Genre" G
    ON T."GenreId" = G."GenreId"
    WHERE G."Name" = 'Jazz';
```

# Find artists whose music are on no playlist

```python
%%sql

SELECT *
    FROM "Playlist";
```

```python
%%sql

SELECT (regexp_split_to_table(R."Name", E'[/,&]')) AS "ArtistName", PT."PlaylistId"
    FROM "Artist" R
    LEFT JOIN "Album" A
    ON R."ArtistId" = A."ArtistId"
    LEFT JOIN "Track" T
    ON A."AlbumId" = T."AlbumId"
    LEFT JOIN "PlaylistTrack" PT
    ON T."TrackId" = PT."TrackId"
    WHERE PT."PlaylistId" IS NULL;
```

Answer: See the table above.
