```python
%%sql

SELECT table_name, table_schema
    FROM information_schema.tables
    WHERE table_schema NOT IN ('pg_catalog', 'information_schema');
```

# Who's the composer of the track 'Cochise'?

```python
%%sql

SELECT *
    FROM information_schema.columns
    WHERE table_name = 'Track';
```

```python
%%sql

SELECT "Name", "Composer"
    FROM "Track"
    WHERE "Name" = 'Cochise';
```

Answer: Audioslave/Chris Cornell

# What's the name of genre for the track 'Cochise'?

```python
%%sql

SELECT *
    FROM information_schema.columns
    WHERE table_name = 'Genre';
```

```python
%%sql

SELECT T."Name" AS "TrackName", G."Name" AS "GenreName"
    FROM "Track" T, "Genre" G
    WHERE T."GenreId" = G."GenreId" AND T."Name" = 'Cochise';
```

Answer: Rock

# What's the top 3 names of genre with the most tracks? How many tracks are there total in that genre?

```python
%%sql

SELECT G."Name" AS "GenreName", COUNT(G."Name") AS "NumOfTracks" 
    FROM "Track" T
    JOIN "Genre" G
    ON T."GenreId" = G."GenreId"
    GROUP BY G."Name"
    ORDER BY Count(G."Name") DESC
    LIMIT 3;
```

Answer: The top 3 names of genre with the most tracks are Rock, Metal, Alternative & Punk, and their tracks are 1259, 342, 265 separately.

# What's the name of the genre with the 3rd most tracks? How many tracks are there total in that genre?

```python
%%sql

SELECT G."Name" AS "GenreName", COUNT(G."Name") AS "NumOfTracks" 
    FROM "Track" T
    JOIN "Genre" G
    ON T."GenreId" = G."GenreId"
    GROUP BY G."Name"
    ORDER BY Count(G."Name") DESC
    LIMIT 1 OFFSET 2;
```

Answer: The name of genre with the 3rd most tracks is Alternative & Punk, and its number of tracks is 265.

# How many tracks does the artist 'Metallica' have?

```python
%%sql

SELECT R."Name" AS "ArtistName", COUNT(R."Name") AS "NumOfTracks"
    FROM "Track" T
    JOIN "Album" A
    ON T."AlbumId" = A."AlbumId"
    JOIN "Artist" R
    ON A."ArtistId" = R."ArtistId"
    GROUP BY R."Name"
    HAVING R."Name" = 'Metallica';
```

Answer: 112

# Which Country buys the most music (invoice)?

```python
%%sql

SELECT *
    FROM information_schema.columns
    WHERE table_name = 'Invoice';
```

```python
%%sql

SELECT "BillingCountry", COUNT("BillingCountry") AS "NumOfInvoices"
    FROM "Invoice"
    GROUP BY "BillingCountry"
    ORDER BY COUNT("BillingCountry") DESC
    LIMIT 1;
```

Answer: USA

# What are the top 5 Genres in the US?

```python
%%sql

SELECT G."Name" AS "GenreName", COUNT(G."Name") AS "NumOfInvoice"
    FROM "InvoiceLine" IL, "Invoice" I, "Track" T, "Genre" G
    WHERE IL."InvoiceId" = I."InvoiceId" 
    AND IL."TrackId" = T."TrackId" 
    AND T."GenreId" = G."GenreId" 
    AND I."BillingCountry" = 'USA'
    GROUP BY G."Name"
    ORDER BY COUNT(G."Name") DESC
    LIMIT 5;
```

Answer: The top 5 genres in the US is Rock, Metal, Latin, Alternative & Punk, Jazz.
